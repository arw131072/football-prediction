```python
import kagglehub
import os
import pandas as pd

path = kagglehub.dataset_download("martj42/international-football-results-from-1872-to-2017")
print(path)
print(os.listdir(path))

df = pd.read_csv(os.path.join(path, "results.csv"))
df.head()
```

```python
print(df.info())
df.head()
df.columns
```

```python
df["tournament"].value_counts().head(20)
```

```python
# convert to home score results
import numpy as np

df["home result"] = np.where(df["home_score"] > df["away_score"], 2,   # home win
               np.where(df["home_score"] == df["away_score"], 1,  # draw
                0)) # home loss
print(df["home result"].value_counts(normalize=True).mul(100).round(2)) # percentage of each result
```

```python
# rank by datetime
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date").reset_index(drop=True)
df[["date","home_team","away_team", "home result"]].head()
```
